```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
#df = pd.DataFrame(np.random.randn(1000,4),columns=['a', 'b', 'c','d'])
%matplotlib inline
```

**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
# Code goes here!
df.plot.scatter(x='c',y='d',figsize=(10,4),color='g',lw=3)
```

** Create a histogram of the 'd' column.**

```python
# Code goes here!
df['d'].plot.hist()
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
# Code goes here!
plt.style.use("bmh")
df['d'].plot.hist(bins=50,alpha=.4)
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
# Code goes here!
df.plot.scatter(x='a',y='b',color='g',s=df["d"]*100)
```

**Recreate the plot below**

** Create a kde plot of the 'd' column **

```python
# Code goes here!
df['d'].plot.kde()
```

**Create a boxplot comparing all columns in the data.**<br>
Hint: use `color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')` and `sym = 'r+'` 

```python
# Code goes here!
colors = dict(boxes='DarkGreen', whiskers='DarkOrange',
              medians='DarkBlue', caps='Gray')
sym = 'r+'

df.plot.box(color=colors, sym=sym)
```

** Create an area plot of all the columns for just the rows up to 20.**

```python
# Code goes here!
df[10:21].plot.area(alpha=.4)

```

**Try moving the legend in the plot above and display it outside the plot, shown below** [Hint](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas) also [gca](https://stackoverflow.com/questions/27221009/fig-gca-vs-fig-add-subplot).....*If `f` is a variable holding a figure, `f.gca()` returns the axes associated with the figure.*

```python
# Code goes here!
df[10:21].plot.area(alpha=.4)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
