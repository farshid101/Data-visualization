```python
import pandas as pd

def calculate_trading_strategy(month=12, first_money_deposit=100, percent_risk=2, trade_per_month=15):
    total = first_money_deposit

    data = {'Month': [], 'Money': [], 'Profit': [], 'Trade_Total': []}

    for i in range(month):
        per_profit = (total * percent_risk / 100)
        month_total = per_profit * trade_per_month
        data['Month'].append(i + 1)
        data['Money'].append(round(total, 2))
        data['Profit'].append(round(per_profit, 2) * trade_per_month)
        data['Trade_Total'].append(round(month_total + total, 2))
        total = total + month_total

    df = pd.DataFrame(data)
    return df

# Get user input
month = int(input("How many months? (default: 12) ") or 12)
first_money_deposit = int(input("How much deposit do you want to start with? (default: 100) ") or 100)
percent_risk = float(input("What is your risk percent? (default: 2) ") or 2)
trade_per_month = int(input("How many trades do you want to do per month? (default: 15) ") or 15)

result_df = calculate_trading_strategy(month, first_money_deposit, percent_risk, trade_per_month)
result_df
```

```python
import matplotlib.pyplot as plt

def plot_trading_strategy(df):
    # Create a larger figure
    plt.figure(figsize=(12, 6))

    # Create a histogram
    plt.bar(df['Month'], df['Trade_Total'])
    plt.xlabel('Month')
    plt.ylabel('Trade_Total')
    plt.title('Trade_Total per Month')

    # Annotate each bar with its value
    for i, value in enumerate(df['Trade_Total']):
        plt.text(df['Month'][i], value + 0.1, str(value), ha='center', va='bottom')

    plt.grid()
    plt.show()

# Use the calculate_trading_strategy function to get the DataFrame
result_df = calculate_trading_strategy(month=36, first_money_deposit=1000, percent_risk=3, trade_per_month=4)

# Use the plot_trading_strategy function to plot the results
plot_trading_strategy(result_df)
result_df
```
